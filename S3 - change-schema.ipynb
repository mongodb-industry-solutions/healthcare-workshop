```python
import pymongo
from pymongo import MongoClient, errors
import os

mongo_uri = os.getenv("MONGO_URI")
client = pymongo.MongoClient(mongo_uri)

# Select the database and collection
db = client['healthcare_workshop']
collection = db['patients']

# Perform the update operation to transform 'address' into 'addresses'
result = collection.update_many(
    {
        "address": { "$exists": True },    # Check if the 'address' field exists
        "addresses": { "$exists": False }  # Ensure the 'addresses' field does not exist
    },
    [
        {
            "$set": {
                "addresses": [
                    "$address"  # Move the existing 'address' object into the 'addresses' array
                ]
            }
        },
        {
            "$unset": "address"  # Remove the old 'address' field
        }
    ]
)

# Output the result of the operation
print(f"Matched {result.matched_count} documents and modified {result.modified_count} documents.")
```
